```python
#4
import numpy as np


class Matrix:

  def __init__(self, n=0, m=0):
        self.setup(n, m)

  def setup(self, n=0, m=0): #initial values at 0
    if n != 0: #if n input doesn't equal 0 (presence of input)
      self.n =n #set class argument to n variable value
    if m != 0: #repeats
      self.m =m
      self.matrix = [[0]*n for i in range(m)] #creates list (matrix) with all values being 0, iterates an m amount of times
  def printmatrix(self):
    print('MATRIX', [self.n], 'by', [self.m])
    if self.matrix and self.matrix[1][1] is not None:
      for row in self.matrix:
        for element in row:#for each element in row in matrix
          print(element, end='')
        print() #tjis is new line after row
    else:
      print("matrix is empty")


  def index(self, *args):
        if len(args) == 2: #if length of arguments has 2 inputs, which will be i, j, then it will
            i, j = args
            return self.get_element(i, j) #calls individual element, and below, individual rows
        elif len(args) == 1:
            i, = args
            return self.get_row(i)

        else:
            raise TypeError("nuh uh") #returns error

  def equalize(self, og):
        if not isinstance(og, Matrix):
            raise ValueError("nuh uh")
        if self.n != og.n or self.m != og.m:
            raise ValueError("nuh uh wrong size. must be equal sized matrixes")
        self.matrix = [row[:] for row in og.matrix]  # copies from matrix to new matrix

  def shape(self):
        return (self.n, self.m)

  def transpose(self):
        t_matrix = [[self.matrix[i][j] for j in range(self.m)] for i in range(self.n)]
        t = Matrix(self.m, self.n)
        t.matrix = t_matrix
        return t

  def row(self, n):
    if not 0 <= n < self.m: #n got to between values of 0(or equal to) and m instance
        raise IndexError("nuh uh")
    return Matrix(matrix=[self.matrix[n]]) #specific row


  def column(self, n): #kinda does the same thing as above, but column wise. same args slightly dif execution
    if not 0 <= n < self.n:
        raise IndexError("nuh uh")
    return Matrix(matrix=[[self.matrix[i][n]] for i in range(self.m)])

  def to_list(self): #returns matrix as list of lists
    return [row[:] for row in self.matrix]


  def block(self, n_0, n_1, m_0, m_1):
    if not (0 <= n_0 < n_1 <= self.n) or not (0 <= m_0 < m_1 <= self.m):
        raise ValueError("nuh uh")
    block_m = [row[n_0:n_1] for row in self.matrix[m_0:m_1]] #
    return Matrix(matrix=block_m)

  def constant(n, m, c):#constant value times columns for blank in range of n
    return [[c] * m for _ in range(n)]

  def zeros(n, m): #does like the same, but with zeroes
    return constant(n, m, 0)

  def ones(n, m): #now with ones
    return constant(n, m, 1)


  def eye(n): #basically creates rref
    return [[1 if i == j else 0 for j in range(n)] for i in range(n)]

  def scalarmul(self, scalar):
    result = Matrix(self.n, self.m)
    for i in range(self.n):
      for j in range(self.m):
        result.matrix[i][j] = self.matrix[i][j] * scalar
    return result



  def add(self, N): #adding
      if self.n != N.n or self.m != N.m:
          raise ValueError("nuh uh")
      return Matrix(self.n, self.m, matrix=[[self.matrix[i][j] + N.matrix[i][j] for j in range(self.n)] for i in range(self.m)])#does the same thing but adding

  def sub(self, N):
      if self.n != N.n or self.m != N.m:
          raise ValueError("nuh uh")
      return Matrix(self.n, self.m, matrix=[[self.matrix[i][j] - N.matrix[i][j] for j in range(self.n)] for i in range(self.m)])#same but subtracting

  def mat_mult(self, N):
      if self.m != N.n:
          raise ValueError("nuhuh")
      return Matrix(N.n, self.m, matrix=[[sum(self.matrix[i][k] * N.matrix[k][j] for k in range(self.n)) for j in range(N.n)] for i in range(self.m)])#same but multiplying matrices

  def element_mult(self, N):
    if self.n != N.n or self.m != N.m:
          raise ValueError("nuhuh")
    return Matrix(self.n, self.m, matrix=[[self.matrix[i][j] * N.matrix[i][j] for j in range(self.n)] for i in range(self.m)])

  def equals(self, N):
    return self.matrix == N.matrix #returns true or false
```

```python
class Matrix:
    def scalarmul(self, scalar):
        result = Matrix(self.rows, self.cols)
        for i in range(self.rows):
            for j in range(self.cols):
                result.matrix[i][j] = self.matrix[i][j] * scalar
        return result
```
