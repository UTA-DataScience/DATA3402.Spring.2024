```python
#3
def histogram(x,n_bins=10,x_min=None,x_max=None):
  if x_min is None:
        x_min = min(x)
  if x_max is None:
        x_max = max(x)

  bin_size = (x_max - x_min) / n_bins
  hist = [0] * n_bins

  bin_edges = [x_min + i * bin_size for i in range(n_bins + 1)]
  for val in x:
        for i in range(n_bins):
            if x_min + i * bin_size <= val < x_min + (i + 1) * bin_size:
                hist[i] += 1
                break

  return hist, bin_edges
```

```python
#4
def draw_histogram(x, n_bins, x_min=None, x_max=None, character="#", max_character_per_line=20):
    hist, bin_edges = histogram(x, n_bins, x_min, x_max)
    max_count = max(hist)
    for i in range(n_bins):
        lower_bound, upper_bound = bin_edges[i], bin_edges[i + 1]
        count = hist[i]
        scaled_count = int(count / max_count * max_character_per_line)
        print(f"[{lower_bound:4.1f}, {upper_bound:4.1f}] : {character * scaled_count}")

```

```python

def generate_function(func, x_min, x_max, N=1000):
    out = []

    func_val = [func(x) for x in range(int(x_min), int(x_max) + 1)]
    func_min, func_max = min(func_val), max(func_val)

    while len(out) < N:
        test_x = random.uniform(x_min, x_max)
        p = random.uniform(func_min, func_max)
        if p <= func(test_x):
            out.append(test_x)

    return out
```
