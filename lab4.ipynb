```python
#9
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        raise NotImplementedError("Area calculation not implemented.")

    def perimeter(self):
        raise NotImplementedError("Perimeter/circumference calculation not implemented.")

    def get_points_on_parameter(self, num_points=16):
        raise NotImplementedError("Parameter points calculation not implemented.")

    def is_point_inside(self, x, y):
        raise NotImplementedError("Point inside check not implemented.")

    def does_overlap(self, other):
        raise NotImplementedError("Overlap check not implemented.")

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def area(self):
        return self.get_length() * self.get_width()

    def perimeter(self):
        return 2 * (self.get_length() + self.get_width())

    def get_points_on_parameter(self, num_points=16):
        points = []
        for i in range(num_points):
            t = i / num_points
            if t < 0.25:
                x_point = self.get_x() + t * self.get_length()
                y_point = self.get_y()
            elif t < 0.5:
                x_point = self.get_x() + self.get_length()
                y_point = self.get_y() + (t - 0.25) * self.get_width()
            elif t < 0.75:
                x_point = self.get_x() + (0.75 - t) * self.get_length()
                y_point = self.get_y() + self.get_width()
            else:
                x_point = self.get_x()
                y_point = self.get_y() + (1 - t) * self.get_width()

            points.append((x_point, y_point))

        return points

    def is_point_inside(self, x, y):
        return (
            self.get_x() <= x <= self.get_x() + self.get_length() and
            self.get_y() <= y <= self.get_y() + self.get_width()
        )

    def does_overlap(self, other):
        return (
            self.get_x() < other.get_x() + other.get_length() and
            self.get_x() + self.get_length() > other.get_x() and
            self.get_y() < other.get_y() + other.get_width() and
            self.get_y() + self.get_width() > other.get_y()
        )

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius

    def get_radius(self):
        return self.__radius

    def area(self):
        return math.pi * self.get_radius() ** 2

    def perimeter(self):
        return 2 * math.pi * self.get_radius()

    def get_points_on_parameter(self, num_points=16):
        points = []
        for i in range(num_points):
            angle = i * (2 * math.pi / num_points)
            x_point = self.get_x() + self.get_radius() * math.cos(angle)
            y_point = self.get_y() + self.get_radius() * math.sin(angle)
            points.append((x_point, y_point))

        return points

    def is_point_inside(self, x, y):
        distance = math.sqrt((x - self.get_x())**2 + (y - self.get_y())**2)
        return distance <= self.get_radius()

    def does_overlap(self, other):
        distance_between_centers = math.sqrt(
            (self.get_x() - other.get_x())**2 + (self.get_y() - other.get_y())**2
        )
        return distance_between_centers < self.get_radius() + other.get_radius()

class Triangle(Shape):
    def __init__(self, base, height, x, y):
        super().__init__(x, y)
        self.__base = base
        self.__height = height

    def get_base(self):
        return self.__base

    def get_height(self):
        return self.__height

    def area(self):
        return 0.5 * self.get_base() * self.get_height()

    def perimeter(self):
        side_c = math.sqrt(self.get_base()**2 + self.get_height()**2)
        return self.get_base() + self.get_height() + side_c

    def get_points_on_parameter(self, num_points=16):
        points = []
        for i in range(num_points):
            t = i / num_points
            if t < 0.33:
                x_point = self.get_x() + t * self.get_base()
                y_point = self.get_y()
            elif t < 0.66:
                x_point = self.get_x() + self.get_base()
                y_point = self.get_y() + (t - 0.33) * self.get_height()
            else:
                x_point = self.get_x() + (1 - t) * self.get_base()
                y_point = self.get_y() + self.get_height()

            points.append((x_point, y_point))

        return points

    def is_point_inside(self, x, y):
        def sign(p1, p2, p3):
            return (p1[0] - p3[0]) * (p2[1] - p3[1]) - (p2[0] - p3[0]) * (p1[1] - p3[1])

        d1 = sign((x, y), (self.get_x(), self.get_y()), (self.get_x() + self.get_base(), self.get_y()))
        d2 = sign((x, y), (self.get_x() + self.get_base(), self.get_y()), (self.get_x(), self.get_y() + self.get_height()))
        d3 = sign((x, y), (self.get_x(), self.get_y() + self.get_height()), (self.get_x(), self.get_y()))

        has_neg = (d1 < 0) or (d2 < 0) or (d3 < 0)
        has_pos = (d1 > 0) or (d2 > 0) or (d3 > 0)

        return not (has_neg and has_pos)

    def does_overlap(self, other):
        vertices_self = [
            (self.get_x(), self.get_y()),
            (self.get_x() + self.get_base(), self.get_y()),
            (self.get_x(), self.get_y() + self.get_height())
        ]

        vertices_other = [
            (other.get_x(), other.get_y()),
            (other.get_x() + other.get_base(), other.get_y()),
            (other.get_x(), other.get_y() + other.get_height())
        ]

        for vertex in vertices_self:
            if other.is_point_inside(vertex[0], vertex[1]):
                return True

        for vertex in vertices_other:
            if self.is_point_inside(vertex[0], vertex[1]):
                return True

        return False

# test
rectangle = Rectangle(length=5, width=3, x=1, y=2)
circle = Circle(radius=4, x=2, y=3)
triangle = Triangle(base=4, height=3, x=1, y=2)

print("Rectangle Does Overlap Circle:", rectangle.does_overlap(circle))
print("Circle Does Overlap Triangle:", circle.does_overlap(triangle))
print("Triangle Does Overlap Rectangle:", triangle.does_overlap(rectangle))
```
